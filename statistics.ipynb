# STATISTICS BASICS

Q 1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Ans--Descriptive and inferential statistics are two fundamental branches of statistics, each serving distinct purposes in data analysis. Below are examples to illustrate their differences and applications.

**Descriptive Statistics Example**:

Descriptive statistics summarize and describe the characteristics of a dataset. For instance, consider the test scores of 1,000 students in a school. Using descriptive statistics, you could calculate:

-  Mean (Average): The average test score is 82.13, providing a central measure of performance.

- Median: The middle score is 84, indicating that half of the students scored above and half below this value.

- Range: The scores range from 45 to 100, showing a spread of 55 points.

- Visualization: A histogram could reveal that most students scored between 70 and 90, forming a bell-shaped distribution.

These statistics help summarize the dataset, making it easier to understand trends and distributions without analyzing individual scores.

**Inferential Statistics Example**:

Inferential statistics use a sample to make predictions or draw conclusions about a larger population. For example, suppose you want to estimate the average height of adult women in a country. Instead of measuring every woman, you could:

- Sample: Measure the heights of 1,000 randomly selected women.

- Confidence Interval: Calculate a 95% confidence interval, such as [155 cm, 163 cm], indicating you are 95% confident the true average height lies within this range.

- Hypothesis Testing: Test whether the average height is significantly different from a global average, using statistical tests like a t-test.

This approach allows you to generalize findings from the sample to the entire population, provided the sample is representative.

Q 2.What is sampling in statistics? Explain the differences between random
and stratified sampling

Ans--**Sampling**:


- Sampling is a statistical technique used to select a portion of a population to represent the entire group. This method allows researchers to draw conclusions about a population without needing to survey every individual, saving time and resources while still obtaining accurate results. The goal is to ensure that the sample accurately reflects the characteristics of the population.

**Random Sampling**

- Definition: Random sampling is a method where each member of the population has an equal chance of being selected. This technique is often used when the population is homogeneous, meaning that individuals have similar characteristics.
Process: Members are chosen entirely by chance, often using random number generators or lottery methods. This approach helps minimize bias and ensures that the sample is representative of the population as a whole.


- Advantages: It is simple to implement and provides a straightforward way to gather data without needing to consider specific subgroups within the population.



**Stratified Sampling**:

- Definition: Stratified sampling involves dividing the population into distinct subgroups, or strata, based on shared characteristics (e.g., age, gender, income). A random sample is then taken from each stratum in proportion to its representation in the population.


- Process: This method ensures that all subgroups are adequately represented in the final sample. For example, if a population consists of 60% females and 40% males, the sample will reflect this ratio.

- Advantages: Stratified sampling provides more accurate and reliable results, especially when the population is heterogeneous. It allows for detailed analysis of specific subgroups, which can be crucial for studies focusing on particular characteristics or behaviors

Q 3.Define mean, median, and mode. Explain why these measures of central
tendency are important.

Ans--Mean:
-  The mean, often referred to as the average, is calculated by adding all the values in a data set and dividing by the number of values. It is sensitive to outliers, which can skew the result. For example, the mean of the numbers 2, 3, and 10 is

(2+3+10)=15


(2+3+10)/3=5.

**Median:**
-  The median is the middle value when a data set is ordered from least to greatest. If there is an even number of observations, the median is the average of the two middle numbers. For instance, in the data set 1, 3, 3, 6, 7, 8, the median is 6. If the data set were 1, 2, 3, 4, 5, 6, the median would be

(3+4)/2=3.5

(3+4)/2=3.5.


**Mode:**:
 - The mode is the value that appears most frequently in a data set. A data set may have one mode, more than one mode (bimodal or multimodal), or no mode at all. For example, in the data set 1, 2, 2, 3, the mode is 2.


**Importance of Measures of Central Tendency**

- Data Summary: These measures provide a quick summary of a data set, allowing for easier interpretation and understanding of the data's overall behavior.

- Comparison: They enable comparisons between different data sets. For example, comparing the average test scores of two classes can help assess performance differences.

- Decision Making: In fields like business, education, and healthcare, understanding the central tendency helps in making informed decisions based on data trends. For instance, knowing the average income in a region can guide economic policies.

- Identifying Trends: These measures help identify trends and patterns within data, which can be crucial for forecasting and strategic planning.


In summary, mean, median, and mode are essential statistical tools that provide valuable insights into data sets, facilitating analysis and decision-making across various fields.



Q 4. Explain skewness and kurtosis. What does a positive skew imply about the
data?

Ans--**Skewness**

- Skewness measures the asymmetry of a distribution. It indicates whether the data leans more to one side of the mean.

- Positive Skewness: The right tail is longer, meaning there are a few unusually high values. The order of central tendencies is: Mode < Median < Mean.

- Negative Skewness: The left tail is longer, indicating a few unusually low values. The order of central tendencies is: Mean < Median < Mode.

- Symmetrical Distribution: Skewness is zero, and the data is evenly distributed around the mean.

**Kurtosis**

- Kurtosis measures the tailedness or peakedness of a distribution compared to a normal distribution.

- Leptokurtic: High kurtosis (>3). The distribution has heavy tails and a sharp peak, indicating more extreme values.

- Mesokurtic: Normal kurtosis (=3). The distribution resembles a normal curve with moderate tails and peak.

- Platykurtic: Low kurtosis (<3). The distribution has light tails and a flatter peak, indicating fewer extreme values.

**A positive skew in data indicates that the majority of values are concentrated on the left side, with a long tail extending to the right, suggesting the presence of outliers or extreme values**

Q 5. Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
# Ans--
# Program to compute Mean, Median, and Mode without external libraries

def calculate_mean(numbers):
    """Return the mean (average) of the list."""
    if not numbers:
        raise ValueError("List is empty. Cannot compute mean.")
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    """Return the median of the list."""
    if not numbers:
        raise ValueError("List is empty. Cannot compute median.")

    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    mid = n // 2

    # If even number of elements, median is average of middle two
    if n % 2 == 0:
        return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2
    else:
        return sorted_nums[mid]

def calculate_mode(numbers):
    """Return the mode(s) of the list. Handles multiple modes."""
    if not numbers:
        raise ValueError("List is empty. Cannot compute mode.")

    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1

    max_freq = max(frequency.values())
    modes = [num for num, freq in frequency.items() if freq == max_freq]

    # If every value appears only once, there is no mode
    if max_freq == 1:
        return None
    return modes

# Given list
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

try:
    mean_value = calculate_mean(numbers)
    median_value = calculate_median(numbers)
    mode_value = calculate_mode(numbers)

    print(f"Mean: {mean_value:.2f}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value if mode_value else 'No mode'}")

except ValueError as e:
    print(e)
```

Q 6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
# Ans--
import numpy as np

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# 1. Covariance matrix
cov_matrix = np.cov(x, y, bias=False)  # bias=False for sample covariance
cov_xy = cov_matrix[0, 1]

# 2. Correlation coefficient
corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]

print(f"Covariance: {cov_xy}")
print(f"Correlation Coefficient: {corr_xy}")
```

Q 7.Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
# Ans--
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Convert to NumPy array for calculations
data_array = np.array(data)

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data_array, 25)
Q3 = np.percentile(data_array, 75)
IQR = Q3 - Q1

# Determine outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data_array[(data_array < lower_bound) | (data_array > upper_bound)]

# Plot the boxplot
plt.figure(figsize=(6, 4))
plt.boxplot(data_array, vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot with Outliers")
plt.ylabel("Values")

# Show the plot
plt.show()

# Print results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Q3 - Q1): {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {list(outliers)}")
```

Q 8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]   
daily_sales = [2200, 2450, 2750, 3200, 4000]

```python
# Ans--
import numpy as np

# Data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

try:
    # Convert to NumPy arrays
    ad_spend_arr = np.array(advertising_spend, dtype=float)
    sales_arr = np.array(daily_sales, dtype=float)

    # Compute covariance matrix
    covariance_matrix = np.cov(ad_spend_arr, sales_arr)
    covariance = covariance_matrix[0, 1]

    # Compute Pearson correlation coefficient
    correlation_matrix = np.corrcoef(ad_spend_arr, sales_arr)
    correlation = correlation_matrix[0, 1]

    # Display results
    print(f"Covariance: {covariance:.2f}")
    print(f"Correlation: {correlation:.4f}")

except Exception as e:
    print(f"Error computing correlation: {e}")
```

Q 9.Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
# Ans--
import matplotlib.pyplot as plt

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Validate data: ensure all values are numeric
if not all(isinstance(score, (int, float)) for score in survey_scores):
    raise ValueError("All survey scores must be numbers.")

# Create histogram
plt.figure(figsize=(8, 5))  # Set figure size
plt.hist(
    survey_scores,
    bins=range(min(survey_scores), max(survey_scores) + 2),  # Bin edges
    edgecolor='black',
    color='skyblue'
)

# Add titles and labels
plt.title("Survey Scores Distribution", fontsize=14)
plt.xlabel("Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Show grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the histogram
plt.show()
```
